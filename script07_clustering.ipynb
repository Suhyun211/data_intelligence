<h1>군집 분석</h1>

```python
# 라이브러리
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
import sklearn
sklearn.__version__
```

```python
# 아이리스 데이터 셋
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
```

# K평균 군집화

기본 군집화

```python
X[:5]
```

```python
from sklearn.cluster import KMeans
f = KMeans(n_clusters=3,random_state=0)
f.fit(X)
```

```python
f.labels_
```

```python
pd.crosstab(f.labels_,y)
```

```python
plt.scatter(X[:,0],X[:,1],c=f.labels_)
plt.xlabel('Sepal Length'); plt.ylabel('Sepal Width')
plt.show()
```

스케일 후 군집화

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
Xscaled = scaler.transform(X)
# 평균 0 분산 1인 형태로 스케일링
```

```python
Xscaled[:5,:]
```

```python
print(X.mean(axis=0), Xscaled.mean(axis=0) )
```

```python
print(X.var(axis=0), Xscaled.var(axis=0) )
```

```python
f2 = KMeans(n_clusters=3,random_state=0)
f2.fit(Xscaled)
pd.crosstab(f2.labels_,y)
```

```python
pd.crosstab(f.labels_,f2.labels_)
```

# 가우시안 혼합 모델

기본 군집화

```python
from sklearn.mixture import GaussianMixture
f = GaussianMixture(n_components=3,random_state=0)
f.fit(X)
```

```python
f.predict(X)
```

```python
pd.crosstab(f.predict(X),y)
```

스케일링 후 군집화

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
Xscaled = scaler.transform(X)
```

```python
f2 = GaussianMixture(n_components=3,random_state=0)
f2.fit(Xscaled)
```

```python
pd.crosstab(f2.predict(Xscaled),y)
```

```python
pd.crosstab(f.predict(X),f2.predict(Xscaled))
```

# 계층적 군집화

기본 군집화

```python
from sklearn.cluster import AgglomerativeClustering
f = AgglomerativeClustering(
    n_clusters=3,
    metric='euclidean',
    linkage='complete'
  )
f.fit(X)
```

```python
f.labels_
```

```python
pd.crosstab(f.labels_,y)
```

코싸인 유사도 이용

```python
from sklearn.cluster import AgglomerativeClustering
f = AgglomerativeClustering(
    n_clusters=3,
    metric='cosine',
    linkage='complete'
  )
f.fit(X)
```

```python
pd.crosstab(f.labels_,y)
```

코싸인 유사도 + Single Linkage

```python
from sklearn.cluster import AgglomerativeClustering
f = AgglomerativeClustering(
    n_clusters=3,
    metric='cosine',
    linkage='single'
  )
f.fit(X)
```

```python
pd.crosstab(f.labels_,y)
```
