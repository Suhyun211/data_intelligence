# 데이터 타입

숫자형 데이터

````python
a = 1
b = 2
c = a+b
print(a,'+',b,'=',c) # 출력함수
````

````python
# 사칙연산: 덧셈, 뺄셈, 곱셈, 나눗셈, 나머지
a = 5
b = 3
print(a+b,a-b,a*b,a/b,a%b)
````

문자열 데이터

````python
a = 'Hello'  # 문자열은 '' 혹은 ""로 표현
print(a, len(a)) # len() 문자열의 길이
# 문자열에서 각 문자 참조
# 순서대로 0, 1, 2, ...; 뒤에서부터 -1, -2, -3 ...
print(a[0],a[1],a[-1])
````

````python
a = 'Hello'
b = 'World'
print(a+b) # 문자열 합치기
````

리스트 데이터

````python
# 리스트: 여러개의 값을 갖고 있는 데이터 타입
# 각 원소는 0, 1, 2... 로 참조 (인덱싱)
# 뒤에서부터는 -1, -2 ...
a = [1,-3.1,'hello',[5,'b']]
print(a[0])
print(a[1])
print(a[3],a[3][0])
````

````python
a.append('added') # 새로운 값의 추가
print(a)
````

딕셔너리 데이터

````python
# 딕셔너리: 키-값 의 쌍을 갖는 데이터 타입
# 변수[키]로 값을 참조
a = {'one':1, 'two':2, 3:'three', 4:4}
print(a['one'],a['two'],a[3],a[4])
````

````python
a['five'] = 'fivefive'  # 새로운 키로 새로운 값을 추가
print(a)
````

# 조건문과 반복문

조건문

````python
a = 10
if a < 5:
  print(a,'is less than 5')
elif a < 10:
  print(a,'is larger or equal to 5 and less than 10')
else:
  print(a,'is larger or equal to 10')
````

````python
b = 'hello'
if b == 'world':
  print('equal')
else:
  print('not equal')
````

반복문

````python
A = [1, 'two',[3,4]]
for a in A:
  print(a)
````

````python
s = 0
for i in range(10):
  s = s+i
print(s)
````

````python
s = 0
for i in range(10,100):
  if i%3 == 0:
    s = s+i
print(s)
````

# 예제

1. 소수 판별

````python
n = 97

k = 0
for i in range(2,n):
  if n%i == 0:
    k = k+1

if k == 0:
  print(n,'is a prime number')
else:
  print(n,'is not a prime number')
````

2. 알파벳 세기

````python
alphabets = 'abcdefghijklmnopqrstuvwxyz'
sentence = 'iloveyousomuchandyoulovemetoo'
count = {} # 빈 딕셔너리
for c in alphabets:  # 각 알파벳의 카운트를 0으로 세팅
  count[c] = 0
for c in sentence:   # 문장에 포함된 알파벳을 1씩 증가
  count[c] = count[c]+1
for c in alphabets:  # 존재하는 알파벳에 대하여 출력
  if count[c]>0:
    print(c,count[c])
````
