```python
# 범주별 평균 계산 (smoker 기준)
MEAN = data[['tip', 'smoker']].groupby('smoker').mean()
smoker_tip_mean = MEAN['tip']['Yes']
non_smoker_tip_mean = MEAN['tip']['No']

# 범주별 분산 계산 (smoker 기준)
VAR = data[['tip', 'smoker']].groupby('smoker').var()
smoker_tip_var = VAR['tip']['Yes']
non_smoker_tip_var = VAR['tip']['No']
```

```python
# 샘플 수 계산
n1 = data['smoker'].value_counts()['Yes']
n2 = data['smoker'].value_counts()['No']

# SMD 계산
d = abs(smoker_tip_mean - non_smoker_tip_mean)
pooled_sd = math.sqrt(((n1 - 1) * smoker_tip_var + (n2 - 1) * non_smoker_tip_var) / (n1 + n2 - 2))
smd = d / pooled_sd

print(f'{smd = :.2f}')
```
